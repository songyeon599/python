```python
l = [3, 5, 9, 1, 2]
def merge_list(l1, l2):
    ml1 = l1 + l2
    ml1.sort()
    return ml1
ml1 = merge_list(l, [2,1])
ml2 = merge_list([6, 9, 4], l)

print(ml1)
print(ml2)
```

```python
nlist = [3, 5, 9, 1, 2]
def get_min_max(nlist):
    max_val = max(nlist)
    min_val = min(nlist)
    nlist.remove(max_val)
    nlist.remove(min_val)
    return max_val, min_val
max_val, min_val = get_min_max(nlist)
print(max_val)
print(min_val)
print(nlist)

    
```
