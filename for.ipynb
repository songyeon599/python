```python
height = int(input('높이를 입력하세요: '))
for i in range(height // 2 + 1):
    print(' ' * (height // 2 - i) + '*' * (2 * i + 1))
for i in range(height // 2):
    print(' ' * (i + 1) + '*' * ((height // 2) - i - i + 1))
```

```python
menu = int(input('메뉴 수를 입력하세요: '))
menu_dict = {}
for i in range(1, menu+1):
    food_name = input(f"{i}번째 추가할 음식명: ")
    price = int(input(f"({food_name} 1입분의 가격: "))
    menu_dict[food_name] = price
print('-' * 14)
for key, value in menu_dict.items():
    print(key, value)
print('-' * 14)

answer = input('메뉴를 입력하세요: ')
if answer in menu_dict:
    print(f"{answer}은(는) {menu_dict[answer]}원 입니다.")
else:
    print(f"{answer}은(는) 메뉴에 없습니다.")
```
