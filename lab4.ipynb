```python
scores = input("영어 점수와 수학 점수를 입력하세요: ")
scores_1 = scores.split(",")
if int(scores_1[0]) + int(scores_1[1]) >= 110:
    if int(scores_1[0]) < 40:
        print("영어 점수 부족.")
    elif int(scores_1[1]) < 40:
        print("수학 점수 부족.")
    else:
        print("합격입니다.")
else:
    print("총합 점수 부족.")
```

```python
중간고사 = int(input("중간고사 점수 입력: "))
기말고사 = int(input("기말고사 점수 입력: "))
def calc_everage(중간, 기말):
    return (중간 + 기말) / 2
def calc_grade(평균):
    if 평균 >= 90:
        return "A"
    elif 90 > 평균 >= 80:
        return "B"
    elif 80 > 평균 >= 70:
        return "C"
    elif 70 > 평균 >= 60:
        return "D"
    else:
        return "F"
중간기말 = calc_everage(중간고사, 기말고사)
print(중간기말)
학점 = calc_grade(중간기말)
print(학점)
```
