```python
r = int(input("반지름을 입력하시오: "))
pi = 3.14
둘레 = 2 * pi * r
넓이 = pi * (r ** 2)
print("원 둘레:", 둘레, "원 넓이:", 넓이)
```

```python
투입한돈 = int(input("투입한 돈을 입력하시오: "))
물건값 = int(input("물건 값을 입력하시오: "))
거스름돈 = 투입한돈-물건값
_500원 = 거스름돈//500
_100원 = 거스름돈%500//100
print("500원짜리:", _500원, "100원짜리:", _100원)
```

```python
fruit = input('과일 이름을 입력하세요: ')
fruit1 = fruit.split()
fruit2 = input('과일 이름을 하나 더 입력하세요: ')
if fruit2 in fruit1:
    print(f'{fruit2}는 리스트에 포함되어 있습니다.')
else:
    print(f'{fruit2}는 리스트에 포함되어 있지 않습니다.')
```
