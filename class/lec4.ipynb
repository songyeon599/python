```python
def number(a, b, c):
    print(a * b * c)
    print(a + b + c)
    listt = [a, b, c]
    listt.sort()
    return listt[0], listt[1], listt[2]
result = number(1 , 10, 5)
print(result)
```

```python
def find_min_value(numbers):
    min_number = min(numbers)
    return min_number
result = find_min_value([5, 2, 8, 1, 9])
print(result)
```

```python
def merge_list(list1, list2):
    list1.extend(list2)
    return list1
result = merge_list([1, 2, 3], [4, 5, 6])
print(result)
```

```python
money = input("금액과 통화명을 입력하세요: ")
money_1 = money.split()
금액 = int(money_1[0])
통화명 = money_1[1]
if 통화명 == "달러":
    금액 *= 1167
elif 통화명 == "엔":
    금액 *= 1.096
elif 통화명 == "유로":
    금액 *= 1268
else:
    금액 *= 171
print(f"{통화명} {금액}원")
```

```python
숫자 = input("세 개의 숫자를 입력하세요: ")
숫자_1 = 숫자.split()
if int(숫자_1[0]) > int(숫자_1[1]) and int(숫자_1[0]) > int(숫자_1[2]):
    print(숫자_1[0])
elif int(숫자_1[1]) > int(숫자_1[0]) and int(숫자_1[1]) > int(숫자_1[2]):
    print(숫자_1[1])
elif int(숫자_1[2]) > int(숫자_1[1]) and int(숫자_1[2]) > int(숫자_1[0]):
    print(숫자_1[2])
```

```python
전화번호 = input("휴대폰 번호를 입력하세요: ")
앞자리 = 전화번호.split('-')[0]
if 앞자리 == "011":
    print("당신은 skt 사용자입니다.")
elif 앞자리 == "016":
    print("당신은 kt 사용자입니다.")
elif 앞자리 == "019":
    print("당신은 lgu 사용자입니다.")
else:
    print("당신은 알수없음 사용자입니다.")
```
