```python
def price(coin):
    _500원 = coin // 500
    _100원 = (coin % 500) // 100
    _50원 = ((coin % 500) % 100) // 50
    _10원 = (((coin % 500) % 100) % 50) // 10
    return _500원, _100원, _50원, _10원
money = int(input('교환할 금액을 입력하시오: '))
_500원, _100원, _50원, _10원 = price(coin=money)
print(f'500원 짜리 동전 {_500원}개 100원 짜리 동전 {_100원}개 50원 짜리 동전 {_50원}개 10원 짜리 동전 {_10원}개')
```
