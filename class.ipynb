```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, number):
        super().__init__(name, age)
        self.number = number

    def introduce(self):
        print(f"저는 {self.name}학생, 학번은 {self.number}입니다.")

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def introduce(self):
        print(f"저는 {self.name}선생님, 담당 과목은 {self.subject}입니다.")

student = Student("임송연", 17, 1318)
student.introduce()

teacher = Teacher("김미래", 30, "음악")
teacher.introduce()
```

```python
class Electronic:
    def __init__(self, company, name):
        self.company = company
        self.name = name

class Phone(Electronic):
    def __init__(self, company, name, phonenumber):
        super().__init__(company, name)
        self.phonenumber = phonenumber

    def call(self, number):
        print(f"{self.phonenumber}에서 {number}로 전화를 겁니다.")

class Laptop(Electronic):
    def __init__(self, company, name, size):
        super().__init__(company, name)
        self.size = size

    def run(self, program):
        print(f"{self.size}의 크기의 노트북에서 {program}를 실행합니다.")

phone = Phone("삼성", "갤럭시 s25", 123)
phone.call(456)

laptop = Laptop("삼성", "갤럭시북", 16)
laptop.run("카카오톡")
```

```python
class Employee:
    def __init__(self, name, pay):
        self.name = name
        self.pay = pay

class Manager(Employee):
    def __init__(self, name, pay, teammate):
        super().__init__(name, pay)
        self.teammate = teammate

    def show_info(self):
        print(f"이름 : {self.name}, 급여 : {self.pay}, 팀원 수 : {self.teammate}")

class Developer(Employee):
    def __init__(self, name, pay, language):
        super().__init__(name, pay)
        self.language = language
    
    def show_info(self):
        print(f"이름 : {self.name}, 급여 : {self.pay}, 언어 : {self.language}")

manager = Manager("직원 1", 200, 3)
manager.show_info()

developer = Developer("직원 2", 400, "프로그래밍 언어") 
developer.show_info()   
```

```python
class GameCharacter:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power

    def attack(self, other):
        other.hp -= self.power
        print(f"{other.name}의 체력이 {other.hp}입니다.")

gamecharacter1 = GameCharacter("캐릭터1", 100, 5)
gamecharacter2 = GameCharacter("캐릭터2", 150, 9)
gamecharacter1.attack(gamecharacter2)
```

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        pass

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed
    
    def introduce(self):
        print("나는 개입니다.")

    def bark(self):
        print(f"{self.name}가 멍멍 짖습니다!")

class Bird(Animal):
    def __init__(self, name, age, wingspan):
        super().__init__(name, age)
        self.wingspan = wingspan

    def introduce(self):
        print("나는 새입니다.")

    def fly(self):
        print(f"{self.name}가 하늘을 납니다!")
```

```python
class Character:
    def __init__(self, name, hp, attack_power):
        self.name = name
        self.hp = hp
        self.attack_power = attack_power

    def is_alive(self):
        if self.hp > 0:
            return True
        else:
            return False
        
    def take_damage(self, amount):
        self.hp -= amount

class Hero(Character):
    def __init__(self, name, hp, attack_power, weapon):
        super().__init__(name, hp, attack_power)
        self.weapon = weapon

    def attack(self, target):
        target.take_damage(self.attack_power)
        print(f"{self.name}이(가) {self.weapon}(으)로 {target.name}(을)를 공격했습니다! 피해:{self.attack_power}")

class Monster(Character):
    def __init__(self, name, hp, attack_power, type):
        super().__init__(name, hp, attack_power)
        self.type = type

    def attack(self, target):
        target.take_damage(self.attack_power)
        print(f"{self.name}이(가) {target.name}(을)를 공격했습니다! 피해:{self.attack_power}")

player1 = Hero("임송연", 100, 10, "총")
monster1 = Monster("고블린", 500, 100, "고블린")
player1.attack(monster1)
print(monster1.hp)
monster1.attack(player1)
print(player1.hp)
print(player1.is_alive())
```

```python
class Counter:
    def __init__(self):
        self.value = 0

    def increase(self):
        self.value += 1

    def decrease(self):
        self.value -= 1

    def get_value(self):
        return self.value
    
count = Counter()
count.increase()
count.increase()
count.increase()
count.decrease()
count.decrease()
print(count.get_value())
```

```python
class ShoppingCart:
    def __init__(self):
        self.list = []

    def add_item(self, name, price):
        self.list.append((name, price))

    def total_price(self):
        self.total = 0
        for name, price in self.list:
            self.total += price

        return self.total
    
    def show_items(self):
        for name, price in self.list:
            print(name, price)

item = ShoppingCart()
item.add_item("상품1", 1000)
item.add_item("상품2", 2000)
item.add_item("상품3", 3000)
print(item.total_price())
print(item.show_items())

```

```python
class User:
    def __init__(self, name):
        self.name = name

    def send_message(self, room, msg):
        room.broadcast(self.name, msg)

class ChatRoom:
    def __init__(self):
        self.loglist = []

    def broadcast(self, sender, msg):
        self.loglist.append(f"{sender}: {msg}")

    def show_log(self):
        for log in self.loglist:
            print(log)

user1 = User("user1")
chatroom = ChatRoom()
user1.send_message(chatroom, "안녕")
chatroom.show_log()
```
