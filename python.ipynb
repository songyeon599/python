```python
# 온라인 쇼핑몰 시뮬레이터

# 상품 데이터베이스 (상품명: [가격, 재고])
product_db = {
    "노트북": [1200000, 5],
    "무선마우스": [25000, 10],
    "USB-C 허브": [40000, 3],
    "키보드": [70000, 7]
}

# 사용자 정보
user_name = input("👤 사용자 이름을 입력하세요: ")

# 장바구니 초기화
cart = {}

print(f"\n🛒 {user_name}님, 쇼핑몰에 오신 걸 환영합니다!\n")

while True:
    print("\n=== 메뉴 ===")
    print("1. 상품 목록 보기")
    print("2. 장바구니에 상품 담기")
    print("3. 장바구니 보기")
    print("4. 결제하기")
    print("5. 종료")

    choice = input("원하는 메뉴 번호를 선택하세요: ")

    if choice == "1":
        print("\n📦 상품 목록:")
        for name, (price, stock) in product_db.items():
            print(f" - {name}: {price}원 (재고: {stock}개)")

    elif choice == "2":
        item = input("상품명을 입력하세요: ")
        if item in product_db:
            if product_db[item][1] > 0:
                quantity = int(input(f"{item} 몇 개 담으시겠습니까? "))
                if quantity <= product_db[item][1]:
                    cart[item] = cart.get(item, 0) + quantity
                    product_db[item][1] -= quantity
                    print(f"{item} {quantity}개 장바구니에 담았습니다.")
                else:
                    print("❗ 재고가 부족합니다.")
            else:
                print("❗ 해당 상품은 품절입니다.")
        else:
            print("❗ 존재하지 않는 상품입니다.")

    elif choice == "3":
        if not cart:
            print("\n🧺 장바구니가 비어 있습니다.")
        else:
            print("\n🧺 장바구니 내용:")
            total = 0
            for item, qty in cart.items():
                price = product_db[item][0]
                subtotal = price * qty
                print(f" - {item}: {qty}개 × {price}원 = {subtotal}원")
                total += subtotal
            print(f"총 결제 금액: {total}원")

    elif choice == "4":
        if not cart:
            print("장바구니가 비어 있어 결제할 수 없습니다.")
        else:
            total = sum(product_db[item][0] * qty for item, qty in cart.items())
            print(f"\n💳 총 결제 금액은 {total}원입니다.")
            confirm = input("결제하시겠습니까? (y/n): ").lower()
            if confirm == "y":
                print("✅ 결제가 완료되었습니다. 감사합니다!")
                cart.clear()
            else:
                print("❌ 결제가 취소되었습니다.")

    elif choice == "5":
        print(f"\n👋 {user_name}님, 이용해 주셔서 감사합니다!")
        break

    else:
        print("❗ 올바른 메뉴 번호를 입력해주세요.")
```

```python
import math

def projectile_motion_no_import(v0, angle_deg):
    g = 9.81  # 중력 가속도
    
    angle_rad = angle_deg * 3.141592653589793 / 180  # 라디안 변환 (math 없이 직접 계산)
    
    # 비행 시간 계산
    t_flight = 2 * v0 * sin(angle_rad) / g
    
    # 최고점 시간과 높이
    t_vertex = v0 * sin(angle_rad) / g
    y_vertex = v0 * sin(angle_rad) * t_vertex - 0.5 * g * t_vertex ** 2
    
    print(f"비행 시간: {t_flight:.2f}초")
    print(f"최고점 높이: {y_vertex:.2f}m")
    
    # 간단한 sin 함수 직접 구현 (테일러 급수 근사)
def sin(x):
    # x는 라디안
    # 테일러 급수로 sin 근사 (5항까지)
    result = 0
    sign = 1
    fact = 1
    power = x
    for i in range(1, 10, 2):
        term = power / fact
        result += sign * term
        sign *= -1
        power *= x * x
        fact *= (i+1) * (i+2)
    return result

def cos(x):
    # 테일러 급수 근사 (5항까지)
    result = 0
    sign = 1
    fact = 1
    power = 1
    for i in range(0, 10, 2):
        term = power / fact
        result += sign * term
        sign *= -1
        power *= x * x
        fact *= (i+1) * (i+2) if i+1 < 10 else 1
    return result

# 사용자 입력
v0 = float(input("초기 속도 (m/s): "))
angle = float(input("발사 각도 (도): "))

projectile_motion_no_import(v0, angle)
```

```python
def vertex(a, b, c):
    x_v = -b / (2*a)
    y_v = a*x_v**2 + b*x_v + c
    return x_v, y_v

def text_parabola(a, b, c):
    x_v, y_v = vertex(a, b, c)
    print(f"함수: {a}x² + {b}x + {c}")
    print(f"꼭짓점 좌표: ({x_v:.2f}, {y_v:.2f})")
    
    # x 범위 설정 (꼭짓점 기준 +-10)
    x_start = int(x_v) - 10
    x_end = int(x_v) + 10
    
    # y값 범위 계산 (x_start ~ x_end)
    y_values = []
    for x in range(x_start, x_end+1):
        y = a*x**2 + b*x + c
        y_values.append(y)
    
    y_min = int(min(y_values)) - 1
    y_max = int(max(y_values)) + 1
    
    # 텍스트 그래프 출력 (y축 위에서부터 출력)
    for y_level in range(y_max, y_min - 1, -1):
        line = ''
        for x in range(x_start, x_end+1):
            y = a*x**2 + b*x + c
            # y_level과 y가 가까우면 '*' 찍기
            if abs(y - y_level) < 0.5:
                line += '*'
            elif y_level == 0:
                line += '-'  # x축
            elif x == 0:
                line += '|'  # y축
            else:
                line += ' '
        print(line)

# 사용자 입력 예시
a = float(input("a값 입력: "))
b = float(input("b값 입력: "))
c = float(input("c값 입력: "))

text_parabola(a, b, c)
```
